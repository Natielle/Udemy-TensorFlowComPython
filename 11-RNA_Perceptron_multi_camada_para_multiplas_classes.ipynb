```python
from sklearn import datasets
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score
```

## Aplicando a rede neural com a função de ativação relu e softmax para prever múltiplas classes 

O objetivo é adivinhar a espécie da planta de acordo com suas características.
Banco de dados da Iris. Pode ser encontrado em: https://archive.ics.uci.edu/ml/datasets/iris.


```python
# Carregando os dados
iris = datasets.load_iris()
```

```python
# Atributos (tamanho das pétalas, etc)
X = iris.data

# Labels (espécie da planta)
Y = iris.target

print("\nLabels: \n", Y)
print("\nAtributos: \n", X)
```

```python
# Como os dados não estão padronizados, vamos fazer isso agora
scaler_x = StandardScaler()
X = scaler_x.fit_transform(X)
X
```

```python
# Observando as dimensões de Y
Y.shape
```

```python
# Porém, precisamos que tenha a dimensão (150,1)
Y = Y.reshape(-1,1)
Y.shape
```

Como temos os valores de Y sendo 1, 2 e 3 precisamos utilizar a conversão do OneHotEncoder para definir as categorias através de zeros e uns. Ele funciona da seguinte forma:

![](oneHotEncoder.jpg)

```python
# Categorizando as categorias de maneira "binária"
onehot = OneHotEncoder()
Y = onehot.fit_transform(Y).toarray()
```

```python
# Analisando o resultado
Y
```

Nota-se de que precisamos realizar essa conversão pois cada neurônio de saída irá receber um valor, que pode ser mais próximo de 1 ou de 0. Para que depois convertemos essas probabilidades para, de fato, zero ou 1.

```python
# Aqui estamos separando 30% do dataset para testarmos e 70% para treinarmos o modelo
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3)

# Analisando as dimensões
print("X_train: ", X_train.shape)
print("X_test: ", X_test.shape)
print("Y_train: ", Y_train.shape)
print("Y_test: ", Y_test.shape)
```

```python
# Definindo a quantidade de neurônios 
neuronios_entrada = X.shape[1] # obtem a quantidade de atributos
print("Quantidade de neurônios de entrada", neuronios_entrada)

# Vamos trabalhar apenas com uma camada oculta por enquanto
neuronios_oculta = int(np.ceil((X.shape[1] + Y.shape[1]) / 2)) # np.ceil arredonda para cima
print("Quantidade de neurônios de oculta", neuronios_oculta)

neuronios_saida = Y.shape[1]
print("Quantidade de neurônios de saida", neuronios_saida)
```

A estrutura que acabamos de calcular é equivalente a que a figura abaixo mostra:
![](estrutura_RNA_iris.png)

```python
# Portanto precisaremos definir os pesos e os bias
W = {'oculta': tf.Variable(tf.random_normal([neuronios_entrada, neuronios_oculta])),
     'saida': tf.Variable(tf.random_normal([neuronios_oculta, neuronios_saida]))}

B = {'oculta': tf.Variable(tf.random_normal([neuronios_oculta])),
     'saida': tf.Variable(tf.random_normal([neuronios_saida]))}
```

```python
# Visualizando os tipos das variáveis
print("\n W: ", W)
print("\n\n type(W['oculta']): ", type(W['oculta']))
print("\n\n W[\'oculta\']: ", W['oculta'])

# Visualizando os tipos das variáveis
print("\n B: ", B)
print("\n\n type(B['oculta']): ", type(B['oculta']))
print("\n\n B[\'oculta\']: ", B['oculta'])
```

```python
# Definindo os placeholders que irão receber os dados
xph = tf.placeholder('float', [None, neuronios_entrada]) # None -> não importa a quantidade de linhas na dimensão 
yph = tf.placeholder('float', [None, neuronios_saida])   # None -> não importa a quantidade de linhas na dimensão 
```

```python
# Processo feed foward
def mlp(x, W, bias):
    # agora vamos fazer o somatório dos inputs multiplicados pelos pesos e ainda assim vamos acrescentar o bias
    camada_oculta = tf.add(tf.matmul(x, W['oculta']), bias['oculta'])
    
    # Vamos utilizar a função de ativação relu -> f(x) = max(0,x)
    camada_oculta_ativacao = tf.nn.relu(camada_oculta)
    
    # agora vamos fazer o somatório dos neurônios da camada oculta multiplicados pelos pesos 
    # e ainda assim vamos acrescentar o seu respectivo bias
    camada_saida = tf.add(tf.matmul(camada_oculta_ativacao, W['saida']), B['saida'])
    # Nota: ainda não aplicamos a função de ativação na camada de saída
    
    return camada_saida
```

```python
# Criando o modelo
modelo = mlp(xph, W, B)
```

```python
# Então vamos calcular o erro
# logits -> previsões do modelo; label -> classes corretas
# E não precisamos aplicar a função softmax depois porque ela já é aplicada aqui
erro = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits = modelo, labels = yph)) 
print("Erro inicial: ", erro) # Podemos confirmar que será um número escalar

# Vamos minimizar os erros com o otimizador Adam ao invés de usar o gradiente descendente 
otimizador = tf.train.AdamOptimizer(learning_rate = 0.0001).minimize(erro)
```

```python
# Vamos implementar manualmente o batch (lotes)
batch_size = 8
batch_total = int(len(X_train) / batch_size)
batch_total
```

```python
# Podemos ver que dentro dos 13 batches totais, vamos ter em torno de 8 registros/exemplos
X_batches = np.array_split(X_train, batch_total)
X_batches
```

```python
# Agora vamos rodar em várias épocas
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for epoca in range(3000):
        erro_medio = 0.0
        batch_total = int(len(X_train) / batch_size)
        X_batches = np.array_split(X_train, batch_total)
        Y_batches = np.array_split(Y_train, batch_total)
        
        # Aqui iremos otimizar os pesos para cada batch
        for i in range(batch_total):
            X_batch, Y_batch = X_batches[i], Y_batches[i]
            _, custo = sess.run([otimizador, erro], feed_dict = {xph: X_batch, yph: Y_batch})
            erro_medio += custo / batch_total
            
        if epoca % 500 == 0:
            print('Época: ' + str((epoca + 1)) + ' erro médio: ' + str(erro_medio))
            
    W_final, B_final = sess.run([W, B])
```

```python
# Podemos ver que os pesos finais serão
print("\nW_final: \n", W_final)
print("\nB_final: \n", B_final)
```

```python
# Agora vamos fazer as previsões na base de dados de teste
previsoes_teste = mlp(xph, W_final, B_final)
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    r1 = sess.run(previsoes_teste, feed_dict = {xph: X_test}) 
    r2 = sess.run(tf.nn.softmax(r1)) # Aplicando a função de ativação softmax para obtermos as probabilidades
    r3 = sess.run(tf.argmax(r2, 1)) # retorna o índice do maior valor
```
